```julia
function is_pythegorean_triple(i,j,k)
    return i^2+j^2==k^2
end
```

```julia
function list_pythegorean_triple(n)
    count = 0
    for i in 1:n, j in 1:n, k in 1:n
        if is_pythegorean_triple(i,j,k)
#             println("$i, $j, $k")
            count += 1
        end
    end
    return count
end
```

```julia
list_pythegorean_triple(100)
```

## Generica types

```julia
struct Point{T} # T is unspecified type  just stored in very little probably 2 bit cpu space
    x::T
    y::T
end
```

```julia
Point{Int}(1,2)
```

```julia
p = Point{Int64}(1,2)
q = Point{Int64}(-1,2)
p.x
```

```julia
function add(p::Point, q::Point)
    x = p.x+q.x
    y = p.y+q.y
    return Point(x,y)
end
```

```julia
print(add(p,q))
```

```julia
supertype(Point{Int})
```
